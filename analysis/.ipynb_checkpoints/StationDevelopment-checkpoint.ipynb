# Population and Development Analysis

Population and development information is important to consider when judging the TOD potential at stations. Populations support the stations and development can help suggest what areas will grow in population, which can predict what stations will get more use in the future. 

The metrics used in this section are population in the areas surrounding stations in 2023, population increase around stations from 2021-2023, and new homes built around stations from 2021-2023. The 'station area' that was used for analysis was a half-mile radius around the station, which is a common area of use when studying TOD potential because it is an area where the station is accessible and walkable to from any area within. 

Data for this section was gathered from querying the census API for relevant variables. The DART system covers four counties in the Dallas-Fort Worth metro area - Dallas, Tarrant, Collin, and Denton counties - so data was collected for all four counties. The surveys used for the data were ACS 2021, 2022, and 2023. Data was gathered at the block group level to allow data to be aggregated at a small-scale level that was needed for the station study areas. 

To get data for the study areas, a half mile buffer geometry was created around each of the stations. A spatial join was then performed between the buffers and the census block group geometries and the data tied to the census block groups was aggregated by the buffer overlaps. 

```python
#| echo: false
#| output: false
dart_stations = gpd.read_file('./data/Rail_Stops.geojson')

print(dart_stations)
```

```python
#| echo: false
#| output: false
lrt_stations = [
    'WESTMORELAND STATION', 'HAMPTON STATION', 'TYLER VERNON STATION', 
    'ZOO STATION', 'UNT DALLAS STATION', 'CAMP WISDOM STATION',
    'LEDBETTER STATION', 'VA MEDICAL CENTER STATION', 'KIEST STATION',
    'ILLINOIS TC/STATION', 'MORRELL STATION', '8TH & CORINTH STATION',
    'CEDARS STATION', 'CONVENTION CENTER STATION', 'Union Station',
    'NORTH CARROLLTON/FRANKFORD STATION', 'TRINITY MILLS RAIL STATION',
    'DOWNTOWN CARROLLTON STATION', 'FARMERS BRANCH STATION',
    'ROYAL LANE STATION', 'WALNUT HILL/DENTON STATION', 'DFW AIRPORT STATION',
    'BELT LINE STATION', 'DALLAS COLLEGE NORTHLAKE CAMPUS STATION',
    'HIDDEN RIDGE STATION', 'IRVING CONVENTION CENTER STATION',
    'LAS COLINAS URBAN CENTER STATION', 'UNIVERSITY OF DALLAS STATION',
    'BACHMAN STATION', 'BURBANK STATION', 'INWOOD/LOVE FIELD STATION',
    'SOUTHWEST MEDICAL DISTRICT/PARKLAND', 'MARKET CENTER STATION',
    'VICTORY STATION', 'WEST END STATION', 'AKARD STATION', 'ST PAUL STATION',
    'PEARL/ARTS DISTRICT STATION', 'DEEP ELLUM STATION', 'BAYLOR STATION',
    'FAIR PARK STATION', 'MLK STATION', 'HATCHER STATION', 'LAWNVIEW STATION',
    'LAKE JUNE STATION', 'BUCKNER STATION', 'CITYPLACE/UPTOWN STATION',
    'SMU/MOCKINGBIRD STATION', 'WHITE ROCK STATION', 'LAKE HIGHLANDS STATION',
    'LBJ / SKILLMAN STATION', 'FOREST / JUPITER STATION',
    'DOWNTOWN GARLAND STATION', 'DOWNTOWN ROWLETT STATION',
    'LOVERS LANE STATION', 'PARK LANE STATION', 'WALNUT HILL STATION',
    'FOREST LN STATION', 'LBJ / CENTRAL STATION', 'SPRING VALLEY STATION',
    'ARAPAHO CENTER STATION', 'GALATYN PARK STATION', 'CITYLINE/BUSH STATION',
    'DOWNTOWN PLANO STATION', 'PARKER ROAD STATION'
]

dart_lrt_stations = dart_stations[dart_stations['stop_name'].isin(lrt_stations)]

print(dart_lrt_stations)
```

```python
#| echo: false
#| output: false
dart_lrt_stations_buffer = dart_lrt_stations.to_crs(epsg=2276)

dart_lrt_stations_buffer['buffer_0.5mile'] = dart_lrt_stations_buffer.geometry.buffer(2640)

dart_lrt_stations_buffer['buffer_0.5mile'] = dart_lrt_stations_buffer['buffer_0.5mile'].to_crs(epsg=4326)

dart_lrt_stations_buffer['geometry'] = dart_lrt_stations_buffer['geometry'].to_crs(epsg=4326)

print(dart_lrt_stations_buffer)
```

```python
#| echo: false
#| output: false
texas = gpd.read_file('./data/texas_blockgroups.shp')

print(texas)
```

```python
#| echo: false
#| output: false
county_fips = ['113', '121', '439', '085']

county_blockgroups = texas[texas['COUNTYFP'].isin(county_fips)]

print(county_blockgroups)
```

```python
#| echo: false
#| output: false
acs23 = c.remote.APIConnection("ACSDT5Y2023")
acs22 = c.remote.APIConnection("ACSDT5Y2022")
acs21 = c.remote.APIConnection("ACSDT5Y2021")

variable = ("NAME", "B25034_001E", "B25034_002E")
counties = "113,121,439,085"
texas = "48"

def query_year_built(acs):
    return acs.query(
        cols=variable,
        geo_unit="block group",
        geo_filter={"state": texas, "county": counties, "tract": "*"}
    )
        
dallas_new_home_2023 = query_year_built(acs23)
dallas_new_home_2022 = query_year_built(acs22)
dallas_new_home_2021 = query_year_built(acs21)
  
print(dallas_new_home_2021)    
print(dallas_new_home_2022)
print(dallas_new_home_2023)
```

```python
#| echo: false
#| output: false
dallas_new_home_2021 = dallas_new_home_2021.rename(columns={
    'B25034_001E': 'B25034_001E_2021'
})

dallas_new_home_2022 = dallas_new_home_2022.rename(columns={
    'B25034_002E': 'B25034_002E_2022'
})

dallas_new_home_2023 = dallas_new_home_2023.rename(columns={
    'B25034_002E': 'B25034_002E_2023'
})

new_homes_21_23 = pd.concat([dallas_new_home_2021, dallas_new_home_2022[['B25034_002E_2022']], dallas_new_home_2023[['B25034_002E_2023']]], axis=1)

print(new_homes_21_23.head(30))

```

```python
#| echo: false
#| output: false
new_homes_21_23['B25034_002E_2023'] = pd.to_numeric(new_homes_21_23['B25034_002E_2023'], errors='coerce')
new_homes_21_23['B25034_002E_2022'] = pd.to_numeric(new_homes_21_23['B25034_002E_2022'], errors='coerce')
new_homes_21_23['B25034_001E_2021'] = pd.to_numeric(new_homes_21_23['B25034_001E_2021'], errors='coerce')

new_homes_21_23['Total_Homes_2023'] = new_homes_21_23['B25034_002E_2023'] + new_homes_21_23['B25034_001E_2021']

new_homes_21_23.rename(columns={'B25034_001E_2021': 'Total_Homes_2021'}, inplace=True)
new_homes_21_23.rename(columns={'B25034_002E_2023': 'New_Homes_2021-2023'}, inplace=True)

new_homes_21_23['GEOID'] = new_homes_21_23['state'].astype(str) + new_homes_21_23['county'].astype(str) + new_homes_21_23['tract'].astype(str) + new_homes_21_23['block group'].astype(str)

print(new_homes_21_23.head(30))
```

```python
#| echo: false
#| output: false
acs23 = c.remote.APIConnection("ACSDT5Y2023")
acs22 = c.remote.APIConnection("ACSDT5Y2022")
acs21 = c.remote.APIConnection("ACSDT5Y2021")

variable = ("NAME", "B01003_001E")
counties = "113,121,439,085"
texas = "48"

def query_pop(acs):
    return acs.query(
        cols=variable,
        geo_unit="block group",
        geo_filter={"state": texas, "county": counties, "tract": "*"}
    )
        
dallas_pop_2023 = query_pop(acs23)
dallas_pop_2022 = query_pop(acs22)
dallas_pop_2021 = query_pop(acs21)
  
print(dallas_pop_2021)    
print(dallas_pop_2022)
print(dallas_pop_2023)
```

```python
#| echo: false
#| output: false
dallas_pop_2021 = dallas_pop_2021.rename(columns={
    'B01003_001E': 'B01003_001E_2021'
})

dallas_pop_2022 = dallas_pop_2022.rename(columns={
    'B01003_001E': 'B01003_001E_2022'
})

dallas_pop_2023 = dallas_pop_2023.rename(columns={
    'B01003_001E': 'B01003_001E_2023'
})

pop_21_23 = pd.concat([dallas_pop_2021, dallas_pop_2022[['B01003_001E_2022']], dallas_pop_2023[['B01003_001E_2023']]], axis=1)

print(pop_21_23.head(30))
```

```python
#| echo: false
#| output: false
pop_21_23['B01003_001E_2023'] = pd.to_numeric(pop_21_23['B01003_001E_2023'], errors='coerce')
pop_21_23['B01003_001E_2022'] = pd.to_numeric(pop_21_23['B01003_001E_2022'], errors='coerce')
pop_21_23['B01003_001E_2021'] = pd.to_numeric(pop_21_23['B01003_001E_2021'], errors='coerce')

pop_21_23['Population_Change_2021_2023'] = pop_21_23['B01003_001E_2023'] - pop_21_23['B01003_001E_2021']

pop_21_23.rename(columns={'B01003_001E_2021': 'Population_2021'}, inplace=True)
pop_21_23.rename(columns={'B01003_001E_2023': 'Population_2023'}, inplace=True)

pop_21_23['GEOID'] = pop_21_23['state'].astype(str) + pop_21_23['county'].astype(str) + pop_21_23['tract'].astype(str) + pop_21_23['block group'].astype(str)

print(pop_21_23.head(30))
```

```python
#| echo: false
#| output: false
blockgroups_newhome_pop = county_blockgroups \
    .merge(pop_21_23[['GEOID', 'NAME', 'Population_2021', 'Population_2023', 'Population_Change_2021_2023']], on='GEOID', how='left') \
    .merge(new_homes_21_23[['GEOID', 'Total_Homes_2021', 'Total_Homes_2023', 'New_Homes_2021-2023']], on='GEOID', how='left')

blockgroups_newhome_pop = gpd.GeoDataFrame(
    blockgroups_newhome_pop[['NAME', 'Population_2021', 'Population_2023', 'Population_Change_2021_2023', 
                             'Total_Homes_2021', 'Total_Homes_2023', 'New_Homes_2021-2023', 'geometry']],
    geometry='geometry'
)

print(blockgroups_newhome_pop)
```

```python
#| echo: false
#| output: false
station_blockgroup_join = gpd.sjoin(dart_lrt_stations_buffer, blockgroups_newhome_pop, how='inner', op='intersects')

station_newhomes_pop = station_blockgroup_join.groupby(['stop_name']).agg({
    'Population_2021': 'sum',
    'Population_2023': 'sum',
    'Population_Change_2021_2023': 'sum',
    'Total_Homes_2021': 'sum',
    'Total_Homes_2023': 'sum',
    'New_Homes_2021-2023': 'sum'
}).reset_index()

station_newhomes_pop['station_buffer'] = station_blockgroup_join.groupby(['stop_name']).first()['buffer_0.5mile'].reset_index(drop=True)

station_newhomes_pop['station_location'] = station_blockgroup_join.groupby(['stop_name']).first()['geometry'].reset_index(drop=True)

station_newhomes_pop = gpd.GeoDataFrame(station_newhomes_pop, geometry='station_buffer')

print(station_newhomes_pop)
```

```python
#| echo: false
#| output: false
train_routes = gpd.read_file('./data/maps/Rail_Routes/Rail_Routes.shp')

lrt_routes = train_routes[~train_routes['route_shor'].isin(['TRE', '425', '620'])]

lrt_routes = lrt_routes.to_crs(epsg=4326)

print(lrt_routes)
```

```python
#| echo: false
#| output: false
station_newhomes_pop['Pop_23'] = station_newhomes_pop['Population_2023'].copy()
station_newhomes_pop['Pop_Change_21_23'] = station_newhomes_pop['Population_Change_2021_2023'].copy()
station_newhomes_pop['New_Homes_21_23'] = station_newhomes_pop['New_Homes_2021-2023'].copy()

scaler = MinMaxScaler()
station_newhomes_pop[['Population_2023', 'Population_Change_2021_2023', 'New_Homes_2021-2023']] = scaler.fit_transform(
    station_newhomes_pop[['Population_2023', 'Population_Change_2021_2023', 'New_Homes_2021-2023']]
)

station_newhomes_pop['Weighted_Score'] = (0.5 * station_newhomes_pop['Population_2023'] + 
                                          0.35 * station_newhomes_pop['Population_Change_2021_2023'] + 
                                          0.25 * station_newhomes_pop['New_Homes_2021-2023'])

station_newhomes_pop['Rank'] = station_newhomes_pop['Weighted_Score'].rank(ascending=False, method='min')

station_newhomes_pop = station_newhomes_pop.set_crs(epsg=4326)

dart_lrt_stations_areas_ranked = station_newhomes_pop.set_geometry('station_location')

dart_lrt_stations_areas_ranked = station_newhomes_pop.sort_values(by='Rank')

dart_lrt_stations_areas_ranked[['stop_name', 'Pop_23', 'Pop_Change_21_23', 'New_Homes_21_23', 'Weighted_Score', 'Rank', 'station_location']]
```

### Population Around Stations

Many of the stations have a sizeable population within a half mile, and this not just at the stations in central, dense areas. There are a good amount of station areas outside of Dallas that have a good sized population, which suggests that there might be TOD opportunities at a good amount of the DART light rail stations based on population metrics. 

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_areas_ranked, x="stop_name", y="Pop_23", color="blue")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("Population Within Half-Mile of Station (2023)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.tight_layout()
plt.show()
```

### Population Change Around Stations 2021-2023

Population change can be a good predictor about what areas might be growing and where TOD focus should be. A good amount of the station areas saw an increase in population between 2021 and 2023. The station areas that increased in population were dispersed across the region rather than being concentrated in specific areas. This is also true of the areas that saw a decrease in population.

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_areas_ranked, x="stop_name", y="Pop_Change_21_23", color="blue")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("Population Change Within Half-Mile of Station (2021-2023)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.tight_layout()
plt.show()
```

### New Homes Built Around Stations 2021-2023

New homes built from 2021-2023 was more concentrated in certain areas than where population increase occurred during the same time. Most of the areas that saw new home development are in centrally located neighborhoods. The reason for lack of development around many of the stations could be due to the car-centric development pattern that is still prevalent in the Dallas area. While DART is looking to promote TOD development in the region, driving is still definitely the primary means of transportation for most people in the region and as such development is not tied to or being focused around transit hubs currently.

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_areas_ranked, x="stop_name", y="New_Homes_21_23", color="blue")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("New Homes Built Within Half-Mile of Station (2021-2023)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("New Homes Built", fontsize=12)
plt.tight_layout()
plt.show()
```

### DART Stations Ranked for TOD Potential Based on Population and Development Metrics

To create a ranking of stations for TOD potential based on population and development metrics, I gave a weighted score to each of the stations based on the three metrics above. Same as with the ridership metrics, I used the MinMaxScaler to normalize each of the three values across the stations and then provided a 'weight' to each of the variables based on how I prioritized the metrics. For this section, I decided that the current population in the surrounding station areas was the most important metric as it best reflects current conditions around the stations. For this sections, population change and new homes development can act as the predictors for future activity in the station areas. They are still important metrics, but more predictive and less indicative of current conditions. For this section, I also decided to weight population change a bit higher than new homes developed because it does not seem that development around transit hubs seems to be a big focus in the region at this time. 

The final interactive map was made using folium and uses graduated colors to distinguish ranking of the stations with the lightest colors being the lowest rank and darkest colors being the highest. Clicking on each of the station symbols opens a pop up that gives an overview of the population and development metrics and the ranking of the station. The light rail routes shape file, which was retrieved from the City of Dallas Open Data Portal, was underlayed on the map to provide more context for the station symbols.

Similar to the ridership metrics map, a good amount of the stations with highest TOD potential based on population and development metrics are located in central areas, especially in downtown Dallas. However, maybe a bit moreso than with the ridership metrics, there seem to be more stations farther out from downtown Dallas with higher rankings, which suggests that there is more TOD potential at stations regionally based on these metrics. 

```python
#| echo: false
max_rank = dart_lrt_stations_areas_ranked['Rank'].max()
min_rank = dart_lrt_stations_areas_ranked['Rank'].min()

colormap = LinearColormap(
    colors=['#095a85', '#109ee9', '#6fc6f5'],
    vmin=min_rank, vmax=max_rank
)

def get_color_for_rank(rank):
    return colormap(rank)

m = folium.Map(location=[32.7767, -96.7970], zoom_start=12)

folium.TileLayer('cartodb positron').add_to(m)

for _, row in lrt_routes.iterrows():
    route_geometry = row['geometry']
    
    if route_geometry.geom_type == 'MultiLineString':
        for line in route_geometry.geoms:
            folium.PolyLine(
                locations=[(coord[1], coord[0]) for coord in line.coords],
                color='#A8A8A8',
                weight=2,
                opacity=0.7
            ).add_to(m)
    else:
        folium.PolyLine(
            locations=[(coord[1], coord[0]) for coord in route_geometry.coords],
            color='#A8A8A8',
            weight=2,
            opacity=0.7
        ).add_to(m)
        
for _, row in dart_lrt_stations_areas_ranked.iterrows():
    lat = row['station_location'].y
    lon = row['station_location'].x
    
    popup_content = f"""
        <strong style="font-size: 16px;">{row['stop_name']}</strong><br>
        <span style="font-size: 14px;">Rank: {row['Rank']}</span><br>
        <span style="font-size: 12px;"><em>Population 2023</em>: {row['Pop_23']}</span><br>
        <span style="font-size: 12px;"><em>Population Change 2021-2023</em>: {row['Pop_Change_21_23']}</span><br>
        <span style="font-size: 12px;"><em>New Homes Built 2021-2023</em>: {row['New_Homes_21_23']}</span>
    """
    
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=get_color_for_rank(row['Rank']),
        fill=True,
        fill_color=get_color_for_rank(row['Rank']),
        fill_opacity=1,
        popup=folium.Popup(popup_content, max_width=300)
    ).add_to(m)

title_html = """
    <div style="position: absolute; 
                top: 10px; 
                right: 10px; 
                font-size: 14px; 
                font-weight: bold; 
                background-color: white; 
                padding: 5px 10px; 
                border-radius: 5px; 
                border: 1px solid #ccc; 
                z-index: 1000;">
        DART Lightrail Stations Ranked by Population and Development Metrics
    </div>
"""

m.get_root().html.add_child(folium.Element(title_html))
m
```
