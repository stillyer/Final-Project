# Ridership and Service Analysis

Ridership and service are important considerations when determining what stations might have strong potential for transit-oriented development. Ridership gauges how many people use the station and may also suggest that they are coming from or using the areas that surround the station. Service - including how many different routes stop at a particular station - can help determine how much activity a station receives and how accessible a station is to riders.

The metrics that were used in this section include average weekday ridership at stations in 2023, the ridership change at stations between 2021-2023, and the average daily stops at each station. Ridership would measure the average amount of people using the station on a given weekday, the change in ridership measures if the station is becoming more or less frequently used, and the average daily stops - which accounts for all different services stopping at the station including light rail, bus, shuttle, etc - measures the amount of transit service provided at the station. 

Ridership data was gathered from the DART Reference Book, which reports on different performance metrics of DART stations and the system as a whole. The station ridership totals are included as a table in the PDF document, so the data was preprocessed in excel first and made into a CSV to import into python for further use. 

Service data was gathered from the General Transit Feed Specification (GTFS), which keeps real-time and static data on different metrics of transit systems around the world including up-to-date operating schedules, arrival and departure times, active routes, and active transit stops. To measure level of service, I decided to calculate how many times different transit routes stop at a station on a given day. To do this I combined different GTFS datasets that includes route information, stop information and location, and arrival and departure times. My focus is specifically on light rail stations, but I also wanted to capture any other services that stop and provide connections at these stations too. Since GTFS uses unique stop ID's for every route (for example, a bus route that stops at a light rail station will have a different stop ID than the stop ID at the station for the light rail route) I created a small buffer around the light rail station stop and did a spatial join to capture all stops that were within the buffer, which I assume as being part of the same station network. GTFS data is captured in weeklong intervals, so once all the stops at each station were counted the total was divided by 7 to get the average daily stop total.

```python
#| echo: false
#| output: false
dart_stations = gpd.read_file('./data/Rail_Stops.geojson')

lrt_stations = [
    'WESTMORELAND STATION', 'HAMPTON STATION', 'TYLER VERNON STATION', 
    'ZOO STATION', 'UNT DALLAS STATION', 'CAMP WISDOM STATION',
    'LEDBETTER STATION', 'VA MEDICAL CENTER STATION', 'KIEST STATION',
    'ILLINOIS TC/STATION', 'MORRELL STATION', '8TH & CORINTH STATION',
    'CEDARS STATION', 'CONVENTION CENTER STATION', 'Union Station',
    'NORTH CARROLLTON/FRANKFORD STATION', 'TRINITY MILLS RAIL STATION',
    'DOWNTOWN CARROLLTON STATION', 'FARMERS BRANCH STATION',
    'ROYAL LANE STATION', 'WALNUT HILL/DENTON STATION', 'DFW AIRPORT STATION',
    'BELT LINE STATION', 'DALLAS COLLEGE NORTHLAKE CAMPUS STATION',
    'HIDDEN RIDGE STATION', 'IRVING CONVENTION CENTER STATION',
    'LAS COLINAS URBAN CENTER STATION', 'UNIVERSITY OF DALLAS STATION',
    'BACHMAN STATION', 'BURBANK STATION', 'INWOOD/LOVE FIELD STATION',
    'SOUTHWEST MEDICAL DISTRICT/PARKLAND', 'MARKET CENTER STATION',
    'VICTORY STATION', 'WEST END STATION', 'AKARD STATION', 'ST PAUL STATION',
    'PEARL/ARTS DISTRICT STATION', 'DEEP ELLUM STATION', 'BAYLOR STATION',
    'FAIR PARK STATION', 'MLK STATION', 'HATCHER STATION', 'LAWNVIEW STATION',
    'LAKE JUNE STATION', 'BUCKNER STATION', 'CITYPLACE/UPTOWN STATION',
    'SMU/MOCKINGBIRD STATION', 'WHITE ROCK STATION', 'LAKE HIGHLANDS STATION',
    'LBJ / SKILLMAN STATION', 'FOREST / JUPITER STATION',
    'DOWNTOWN GARLAND STATION', 'DOWNTOWN ROWLETT STATION',
    'LOVERS LANE STATION', 'PARK LANE STATION', 'WALNUT HILL STATION',
    'FOREST LN STATION', 'LBJ / CENTRAL STATION', 'SPRING VALLEY STATION',
    'ARAPAHO CENTER STATION', 'GALATYN PARK STATION', 'CITYLINE/BUSH STATION',
    'DOWNTOWN PLANO STATION', 'PARKER ROAD STATION'
]

dart_lrt_stations = dart_stations[dart_stations['stop_name'].isin(lrt_stations)]

print(dart_lrt_stations)
```

```python
#| echo: false
#| output: false
stops = pd.read_csv('./data/ridership/stops.txt')
stop_times = pd.read_csv('./data/ridership/stop_times.txt')
routes = pd.read_csv('./data/ridership/routes.txt')
trips = pd.read_csv('./data/ridership/trips.txt')
calendar = pd.read_csv('./data/ridership/calendar.txt')

station_activity = (
    stop_times
    .merge(trips, on='trip_id')
    .merge(stops, on='stop_id')
    .merge(routes, on='route_id')
)

cols = ['route_desc', 'route_type', 'trip_id', 'service_id']
lightrailroutes = ['25823', '25824', '25825', '25826']
print(station_activity[cols][station_activity['route_id'].isin([25823, 25824, 25825, 25826])])
```

```python
#| echo: false
#| output: false
geometry = [Point(xy) for xy in zip(station_activity['stop_lon'], station_activity['stop_lat'])]

station_activity_geo = gpd.GeoDataFrame(station_activity, geometry=geometry)

station_activity_geo.set_crs('EPSG:4326', inplace=True)

print(station_activity_geo)
```

```python
#| echo: false
#| output: false
dart_lrt_stations = gpd.GeoDataFrame(dart_lrt_stations, geometry=dart_lrt_stations['geometry'], crs="EPSG:4326")

station_activity_geo = station_activity_geo.to_crs(epsg=2276)
dart_lrt_stations = dart_lrt_stations.to_crs(epsg=2276)

dart_lrt_stations['stops_area'] = dart_lrt_stations.geometry.buffer(500)
dart_lrt_stations = dart_lrt_stations.set_geometry('stops_area')

stops_station = gpd.sjoin(station_activity_geo, dart_lrt_stations[['stop_name', 'stops_area']], predicate='within')

stops_by_station = stops_station.groupby('stop_name_right').size().reset_index(name='stop_count')

stops_by_station.rename(columns={'stop_name_right': 'stop_name'}, inplace=True)

dart_lrt_stations_stops = dart_lrt_stations.merge(stops_by_station, on='stop_name', how='left').fillna({'stop_count': 0})

dart_lrt_stations_stops['average_daily_stops'] = (dart_lrt_stations_stops['stop_count'] / 7).round()

print(dart_lrt_stations_stops)
```

```python
#| echo: false
#| output: false
dart_lrt_stations_merged = pd.merge(dart_lrt_stations_stops, ridership, left_on='stop_name', right_on='Station', how='inner')

dart_lrt_stations_final = dart_lrt_stations_merged[['stop_name', 'Avg. Weekday Ridership (FY23)', 'Ridership_Change_21-23', 'stop_count', 
                      'average_daily_stops', 'geometry']]

print(dart_lrt_stations_final)
```

```python
#| echo: false
#| output: false
train_routes = gpd.read_file('./data/maps/Rail_Routes/Rail_Routes.shp')

lrt_routes = train_routes[~train_routes['route_shor'].isin(['TRE', '425', '620'])]

lrt_routes = lrt_routes.to_crs(epsg=4326)

print(lrt_routes)
```

```python
#| echo: false
#| output: false
dart_lrt_stations_final['Avg_Weekday_Ridership_23'] = dart_lrt_stations_final['Avg. Weekday Ridership (FY23)'].copy()
dart_lrt_stations_final['Ridership_Change_21_23'] = dart_lrt_stations_final['Ridership_Change_21-23'].copy()
dart_lrt_stations_final['Avg_Daily_Stops'] = dart_lrt_stations_final['average_daily_stops'].copy()

scaler = MinMaxScaler()
dart_lrt_stations_final[['Avg. Weekday Ridership (FY23)', 'Ridership_Change_21-23', 'average_daily_stops']] = scaler.fit_transform(
    dart_lrt_stations_final[['Avg. Weekday Ridership (FY23)', 'Ridership_Change_21-23', 'average_daily_stops']]
)

dart_lrt_stations_final['Weighted_Score'] = (0.5 * dart_lrt_stations_final['Avg. Weekday Ridership (FY23)'] + 
                                              0.25 * dart_lrt_stations_final['Ridership_Change_21-23'] + 
                                              0.25 * dart_lrt_stations_final['average_daily_stops'])

dart_lrt_stations_final['Rank'] = dart_lrt_stations_final['Weighted_Score'].rank(ascending=False, method='min')

dart_lrt_stations_ranked = dart_lrt_stations_final.sort_values(by='Rank')

dart_lrt_stations_ranked = dart_lrt_stations_ranked.set_geometry('geometry')
dart_lrt_stations_ranked = dart_lrt_stations_ranked.to_crs(epsg=4326)

dart_lrt_stations_ranked[['stop_name', 
                          'Avg_Weekday_Ridership_23', 'Ridership_Change_21_23', 
                          'Avg_Daily_Stops', 'Weighted_Score', 'Rank', 'geometry']]
```

### Average Weekday Ridership By Station

As might be expected, many of the stations with the highest ridership are located in central, urban locations, such as in downtown Dallas or in the downtowns of neighboring cities such as Plano. Other stations with high ridership often have a nearby anchor institution, such as a university or hospital, or are near denser residential areas that may have more people using the system to commute to work.

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_ranked, x="stop_name", y="Avg_Weekday_Ridership_23", color="red")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("Average Weekday Ridership by Station (2023)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Average Weekday Ridership", fontsize=12)
plt.tight_layout()
plt.show()
```

### Ridership Change by Station 2021-2023

Ridership numbers have gone up at many of DART's lightrail stations since 2021, with the exception of Burbank Station which had a decrease in ridership. Similar with ridership totals, the stations that saw the biggest increase in ridership over the past two years were often located in central locations or closely linked to significant institutions. The increase could be a return to pre-pandemic activity as we get farther away from the pandemic, but still suggests an overall increase in activity around downtown and landmark locations.

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_ranked, x="stop_name", y="Ridership_Change_21_23", color="red")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("Ridership Change by Station (2021-2023)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Ridership Change", fontsize=12)
plt.tight_layout()
plt.show()
```

### Average Daily Stops by Station

The stations with the most service and daily stops are again oftentimes located in central locations. Many of these stations are connected to multiple, and sometimes all four, of the system's light rail lines as well as connecting bus routes and shuttle service. However, some stations seem to receive more service than their ridership numbers might warrant. For example, Downtown Garland receives a good amount of service, but low ridership compared to many of the other stations in the system. This seems to be a trend for a good amount of stations outside of the main urban centers, but might suggest the potential for future development around these stations since the system could be able to support an increase in riders.

```python
#| echo: false
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=dart_lrt_stations_ranked, x="stop_name", y="Avg_Daily_Stops", color="red")
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.title("Average Daily Stops by Station", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Average Daily Stops", fontsize=12)
plt.tight_layout()
plt.show()
```

### DART Stations Ranked for TOD Potential Based on Ridership Metrics

To create a ranking of stations for TOD Potential based on ridership metrics, I gave a weighted score to each of the stations based on the three metrics above. I used the MinMaxScaler to normalize each of the three values across the stations and then provided a 'weight' to each of the variables based on how I prioritized the metrics. I decided that the current ridership totals were most important as they best reflect current use of the stations. Ridership change and daily stops were weighted the same. They could be good predictors of future use at the stations and what stations could support an increase in ridership, but are more predictive and less concrete than current ridership. 

The final interactive map was made using folium and uses graduated colors to distinguish ranking of the stations with the lightest colors being the lowest rank and darkest colors being the highest. Clicking on each of the station symbols opens a pop up that gives an overview of the ridership metrics and the ranking of the station. The light rail routes shape file, which was retrieved from the City of Dallas Open Data Portal, was underlayed on the map to provide more context for the station symbols.

The map highlights that many of the stations with highest TOD potential based on ridership metrics are located in central areas. A good amount of these stations are in downtown Dallas, but some are also in neighboring cities such as Plano and Garland or at the ends of the routes. 

```python
#| echo: false
max_rank = dart_lrt_stations_ranked['Rank'].max()
min_rank = dart_lrt_stations_ranked['Rank'].min()

colormap = LinearColormap(
    colors=['#580e0e', '#dc2323', '#f1a7a7'],
    vmin=min_rank, vmax=max_rank
)

def get_color_for_rank(rank):
    return colormap(rank)

m = folium.Map(location=[32.7767, -96.7970], zoom_start=12)

folium.TileLayer('cartodb positron').add_to(m)

for _, row in lrt_routes.iterrows():
    route_geometry = row['geometry']
    
    if route_geometry.geom_type == 'MultiLineString':
        for line in route_geometry.geoms:
            folium.PolyLine(
                locations=[(coord[1], coord[0]) for coord in line.coords],
                color='#A8A8A8',
                weight=2,
                opacity=0.7
            ).add_to(m)
    else:
        folium.PolyLine(
            locations=[(coord[1], coord[0]) for coord in route_geometry.coords],
            color='#A8A8A8',
            weight=2,
            opacity=0.7
        ).add_to(m)
        
for _, row in dart_lrt_stations_ranked.iterrows():
    lat = row['geometry'].y
    lon = row['geometry'].x
    
    popup_content = f"""
        <strong style="font-size: 16px;">{row['stop_name']}</strong><br>
        <span style="font-size: 14px;">Rank: {row['Rank']}</span><br>
        <span style="font-size: 12px;"><em>Avg. Weekday Ridership 2023 (# of riders)</em>: {row['Avg_Weekday_Ridership_23']}</span><br>
        <span style="font-size: 12px;"><em>Ridership Change 2021-2023 (# of riders)</em>: {row['Ridership_Change_21_23']}</span><br>
        <span style="font-size: 12px;"><em>Average Daily Service (# of stops)</em>: {row['Avg_Daily_Stops']}</span>
    """
    
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=get_color_for_rank(row['Rank']),
        fill=True,
        fill_color=get_color_for_rank(row['Rank']),
        fill_opacity=1,
        popup=folium.Popup(popup_content, max_width=300)
    ).add_to(m)

title_html = """
    <div style="position: absolute; 
                top: 10px; 
                right: 10px; 
                font-size: 14px; 
                font-weight: bold; 
                background-color: white; 
                padding: 5px 10px; 
                border-radius: 5px; 
                border: 1px solid #ccc; 
                z-index: 1000;">
        DART Lightrail Stations Ranked by Ridership and Service Metrics
    </div>
"""

m.get_root().html.add_child(folium.Element(title_html))
m
```
